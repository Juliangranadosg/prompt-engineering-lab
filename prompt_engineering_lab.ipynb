Herlper Functions

```python
def call_openai(prompt, system_message="You are a helpful assistant.", temperature=0.7, max_tokens=None, seed=None):
    """Helper function to call OpenAI API with specified parameters"""
    try:
        params = {
            "model": MODEL,
            "messages": [
                {"role": "system", "content": system_message},
                {"role": "user", "content": prompt}
            ],
            "temperature": temperature
        }
        
        if max_tokens:
            params["max_tokens"] = max_tokens
        if seed is not None:
            params["seed"] = seed
            
        response = client.chat.completions.create(**params)
        return response.choices[0].message.content
    except Exception as e:
        return f"Error: {str(e)}"

def display_response(title, response, params=None):
    """Display a response with formatting"""
    print(f"\n{'='*60}")
    print(f"📝 {title}")
    if params:
        print(f"Parameters: {params}")
    print(f"{'='*60}")
    print(response)
    print(f"{'='*60}\n")

def count_tokens_approx(text):
    """Approximate token count (roughly 4 characters per token)"""
    return len(text) // 4

print("✅ Helper functions loaded!")
```

create initial prompts

```python
# -----------------------------
# Task 1: Sentiment Analysis
# -----------------------------
sentiment_prompt_v1 = """
Classify the sentiment of the following customer message as Positive, Negative, or Neutral.

Message: "I love this product! It's exactly what I needed."

Answer:
"""

result = call_openai(sentiment_prompt_v1)
print("Sentiment Analysis Result:")
print(result)


# -----------------------------
# Task 2: Product Description
# -----------------------------
product_prompt_v1 = """
Write a short product description for the following item:

Product: Wireless mouse
Price: $29.99

Description:
"""

result = call_openai(product_prompt_v1)
print("\nProduct Description Result:")
print(result)


# -----------------------------
# Task 3: Data Extraction
# -----------------------------
extraction_prompt_v1 = """
Extract key information from the following customer feedback.

Feedback: "I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."

Return the extracted information:
"""

result = call_openai(extraction_prompt_v1)
print("\nData Extraction Result:")
print(result)
```

Various runs

```python
num_runs = 15

all_results = {
    "sentiment": [],
    "product": [],
    "extraction": []
}

for i in range(num_runs):
    print(f"\n--- Run {i+1} ---")

    # Sentiment Analysis
    sentiment_res = call_openai(sentiment_prompt_v1)
    print(f"Sentiment: {sentiment_res}")
    all_results["sentiment"].append(sentiment_res)

    # Product Description
    product_res = call_openai(product_prompt_v1)
    print(f"Product: {product_res[:60]}...")  # shorten output
    all_results["product"].append(product_res)

    # Data Extraction
    extraction_res = call_openai(extraction_prompt_v1)
    print(f"Extraction: {extraction_res}")
    all_results["extraction"].append(extraction_res)
```
